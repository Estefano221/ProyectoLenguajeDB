# [dbo].[ActualizarCostoItinerario]

```sql
/****** Object:  StoredProcedure [dbo].[ActualizarCostoItinerario]    Script Date: 11/8/2023 06:08:53 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ActualizarCostoItinerario]
    @p_id_itinerario INT,
    @p_nuevo_costo DECIMAL(20, 2)
AS
BEGIN
    UPDATE Itinerario
    SET costo_total = @p_nuevo_costo
    WHERE id_itinerario = @p_id_itinerario;
END;
GO
```

# [dbo].[ObtenerCostoTotalViaje]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerCostoTotalViaje]    Script Date: 11/8/2023 06:08:53 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ObtenerCostoTotalViaje]
    @p_id_viaje INT,
    @p_costo_total DECIMAL(20, 2) OUTPUT
AS
BEGIN
    SELECT @p_costo_total = costo_total FROM Viaje WHERE id_viaje = @p_id_viaje;
END;
GO
```

# [dbo].[ObtenerDetallesViaje]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerDetallesViaje]    Script Date: 11/8/2023 06:08:53 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ObtenerDetallesViaje]
    @p_id_viaje INT
AS
BEGIN
    SELECT * FROM Viaje WHERE id_viaje = @p_id_viaje;
    SELECT * FROM Itinerario WHERE id_viaje = @p_id_viaje;
    SELECT * FROM Hospedaje WHERE id_viaje = @p_id_viaje;
    SELECT * FROM Vuelo WHERE id_viaje = @p_id_viaje;
    SELECT * FROM Reservas WHERE id_viaje = @p_id_viaje;
END;
GO
```
